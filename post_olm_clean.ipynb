```python
import json
import re
import time
```

```python
input_json = "/content/output_f9f8c053bf41ce2d60f37de1ff7df79e4cbaec4c.jsonl"

output_file = "blackwell_olmocr.txt"
```

```python
def clean_olmocr_artifacts(text):
    text = re.sub(r'<[^>]+>', '', text)

    text = re.sub(r'^\s*\d+\s*$', '', text, flags=re.MULTILINE)

    text = text.replace('\n', ' ')

    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
input_2 = "/content/output_078e337c4d74a139d43eff836fbc9326012604ed.jsonl"
output2 = 'carlette_parker.txt'


full_transcript = []
with open(input_2, "r") as f:
    for line in f:
        data = json.loads(line)
        #print(data.keys())
        metadata = data['metadata']


        full_transcript.append(clean_olmocr_artifacts(data["text"]))

final_transcript = " ".join(full_transcript)

with open(output2, "w") as file:
    file.write(final_transcript)

```
